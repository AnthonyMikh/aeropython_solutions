```python
N = 50
rect = Rect(-2.0, 2.0, -1.0, 1.0)
grid = rect.make_grid(N)

source = Source(-1.0, 0.0, 5.0)
src_vel = source.velocity(grid)


#stream_params = {'density': 2, 'linewidth': 1, 'arrowsize': 2, 'arrowstyle': '->'}
#scatter_params = {'color': '#CD2305', 's': 80, 'marker': 'o', 'linewidth': 0}

size = 10
setup_fig(rect, size)

plt.streamplot(*grid, *src_vel, **stream_params)
plt.scatter(*source.coords, **scatter_params)
```

```python
sink = Source(1.0, 0.0, -5.0)
sink_vel = sink.velocity(grid)

setup_fig(rect, size)
plt.streamplot(*grid, *sink_vel, **stream_params)
plt.scatter(*sink.coords, **scatter_params)
```

```python
u = src_vel[0] + sink_vel[0]
v = src_vel[1] + sink_vel[1]

setup_fig(rect, size)
plt.streamplot(*grid, u, v, **stream_params)
plt.scatter([source.coords[0], sink.coords[0]], [source.coords[1], sink.coords[1]], **scatter_params)
```

```python
src_pot = source.potential(grid)
sink_pot = sink.potential(grid)
pot = src_pot + sink_pot
plt.contour(*grid, pot, 20)
```
